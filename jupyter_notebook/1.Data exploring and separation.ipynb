# function to get categories

```python
# kw_index_list = []
def get_category(data , column , keywords ,kw_index_list):
        for i in range(len(data)):
            check_list = data.loc[i, str(column)]
            for j in range(len(check_list)):
                if check_list[j] in keywords:
                    kw_index_list.append(i)
                    break
        return kw_index_list
```

# check function

```python
def check_func(data , column , bus_list , keywords):
        true_data = pd.DataFrame(columns=data.columns)
        for i in range(len(data)):
            if str(data.loc[i, column]) in bus_list:
                true_data.loc[i, :] = data.loc[i, :]

        true_data.reset_index(drop=True, inplace=True)

        check = {"True": []}
        for i in range(len(true_data)):
            category = true_data.loc[i, "categories"]
            for j in range(len(category)):
                if category[j] in keywords:
                    check["True"].append(i)
                    break
        return len(check["True"])
```
